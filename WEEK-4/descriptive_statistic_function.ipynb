```python
# ---------------------------------
# Descriptive Statistic Functions
# ---------------------------------

import pandas as pd

# Create Data
data = {
    "Ages": [22, 25, 47, 35, 46, 50, 28, 33, 39, 41],
    "Salaries": [5000, 3200, 5500, 4000, 5200, 6000, 3500, 3700, 4500, 4800]
}

# Create DataFrame
df = pd.DataFrame(data)

# Summary Statistics
print("Summary Statistics:\n", df.describe())

# Mean
print("\nMean:\n", df.mean())

# Median
print("\nMedian:\n", df.median())

# Mode
print("\nMode:\n", df.mode())

# Minimum Value
print("\nMinimum Value:\n", df.min())

# Maximum Value
print("\nMaximum Value:\n", df.max())

# Standard Deviation
print("\nStandard Deviation:\n", df.std())

# Variance
print("\nVariance:\n", df.var())

# Count
print("\nCount:\n", df.count())

# Skewness
print("\nSkewness:\n", df.skew())

# Kurtosis
print("\nKurtosis:\n", df.kurt())
```
